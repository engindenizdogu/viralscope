```python
import os
import pandas as pd

# Verify current working directory
print(f"Current working directory: {os.getcwd()}")
```

```python
#FILE_NAME = '_raw_df_channels'
#FILE_NAME = '_raw_df_timeseries'
#FILE_NAME = 'df_channels_en'
#FILE_NAME = 'df_timeseries_en'
FILE_NAME = 'num_comments'
PATH = f'RawData/{FILE_NAME}' + '.tsv.gz'

# Read and display the first few rows
SAMPLE_SIZE = 50_000 # target sample size
CHUNK_SIZE = 1_000_000 # chunk size for reading large files (do not read all into memory)
df = pd.read_csv(PATH, sep='\t', compression='gzip', nrows=CHUNK_SIZE)
df.head()
```

```python
# Display column information
print("\nDataFrame Info:")
df.info()
```

```python
# Sample some of the rows from the chunked dataframe
df_sample = df.sample(n=SAMPLE_SIZE, random_state=42).reset_index(drop=True)
print(f"Sampled dataset shape: {df_sample.shape}")
df_sample.head()
```

```python
# Write to .csv file
df_sample.to_csv(f'SampleData/{FILE_NAME}' + '_sample.csv', index=False)
```
